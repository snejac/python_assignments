7. ### Load the Iris dataset (using pandas) into a Pandas DataFrame and perform data cleaning operations such as removing duplicates, handling missing values, and converting data types.Provide code and explanations for each step.

```python
import os

os.chdir(r"C:\Users\sneja\anaconda3\Lib\site-packages\bokeh_sampledata\_data")
print(os.getcwd())
```

```python
#Loading the Iris dataset (using pandas) into a Pandas DataFrame

import pandas as pd

df = pd.read_csv(r"iris.csv")
df
```

### Cleaning Operation

```python
# Checking for missing values

print("\nMissing values before handling:")
print(df.isnull().sum())
```

```python
# Checking anf Handling for missing values (Option-1: fill with 0)

if df.isnull().values.any():
    print("There are missing values")
   
    df.fillna(0, inplace=True)    # Handling missing values,fill with 0
else:
    print("No missing values")
```

```python
# Checking anf Handling for missing values (Option-2: Fill with the mean (for numerical columns))

for column in ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']:
    df[column].fillna(df[column].mean(), inplace=True )

print(df.isnull().sum())
```

```python
# No. of Duplicate Rows

df[df.duplicated()]
```

```python
# Removing Duplicates

df.drop_duplicates(inplace=True)
print("\nNumber of rows after removing duplicates:", len(df))
```

```python
# Checking Data Types

df.info()
```

8. ###  Using the Iris dataset (using pandas), create the following visualizations using Matplotlib and Seaborn: a histogram, a bar plot, and a scatter plot. Ensure to label the axes and provide titles for each plot.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

#### HISTOGRAM

```python
# Histogram of Sepal Length

plt.figure(figsize=(4, 3))
sns.histplot(df['sepal_length'], bins=10, kde=True)
plt.title('Distribution of Sepal Lengths')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

#### BAR PLOT

```python
# Bar Plot of Average Petal Width by Species

plt.figure(figsize = (4,3))
sns.barplot (x="species", y='petal_width',data=df)
plt.title ("Average Petal Width by Species")
plt.xlabel ("Species")
plt.ylabel ("Average Petal Width")
plt.show()
```

#### SCATTER PLOT

```python
# Scatter Plot of Sepal Length vs Sepal Width

plt.figure(figsize = (4,3))
sns.scatterplot (x = 'sepal_length', y = 'sepal_width', hue = "species", data = df)
plt.title("Sepal Length Vs Sepal Width")
plt.xlabel("Sepal Length")
plt.ylabel('Sepal Width')
plt.legend(title = 'Species')
plt.show()
```

### Data set from sklearn (scikit-learn)

```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
 #Loading the Iris dataset

iris = load_iris() 
```

```python
#Convert to a pandas DataFrame

df = pd.DataFrame (data = iris.data, columns = iris.feature_names )
df ['target'] = iris.target
```

```python
df.head(2)
```

```python
plt.figure(figsize = (4,3))
sns.scatterplot (x = 'sepal length (cm)', y = 'sepal width (cm)', hue = "target", data = df)
plt.title("Sepal Length Vs Sepal Width")
plt.xlabel("Sepal Length")
plt.ylabel('Sepal Width')
plt.legend(title = 'Species')
plt.show()
```
